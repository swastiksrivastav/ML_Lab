KNN

```python
# Importing required libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Load the dataset (assuming it's in a CSV file named 'dataset.csv')
df = pd.read_csv("/content/sample_data/diabetes2.csv")

# Display the DataFrame
print("Dataset:")
df.head()

```

```python

# Plotting a scatter plot (assuming two features for simplicity)
plt.scatter(df['Insulin'], df['BMI'], c=df['Outcome'], cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Scatter Plot of Data')
plt.show()

# Splitting the dataset into features (X) and target variable (y)
X = df.drop('Outcome', axis=1)
y = df['Outcome']

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.45, random_state=42)

# Instantiating the KNN classifier
k = 10 # Assuming k=5 for the number of neighbors
knn_classifier = KNeighborsClassifier(n_neighbors=k)

# Training the classifier
knn_classifier.fit(X_train, y_train)

# Making predictions on the test set
y_pred = knn_classifier.predict(X_test)

# Calculating the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

SVM


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Load the dataset (assuming it's in a CSV file named 'dataset.csv')
df = pd.read_csv("/content/sample_data/Iris.csv")

# Display the first few rows of the DataFrame and set the target variable
print("Dataset:")
print(df.head())
```

```python
class_mapping = {'Iris-setosa': 0, 'Iris-versicolor': 1, 'Iris-virginica': 2}
df['target_numeric'] = df['Species'].map(class_mapping)


plt.scatter(df['SepalWidthCm'], df['PetalWidthCm'], c=df['target_numeric'], cmap='viridis')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Scatter Plot of Iris Data')
plt.show()

# Splitting the dataset into features (X) and target variable (y)
X = df.drop('Species', axis=1)
y = df['Species']

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.45, random_state=50)

# Instantiating the SVM classifier
svm_classifier = SVC(kernel='linear')  # Using linear kernel for simplicity

# Training the classifier
svm_classifier.fit(X_train, y_train)

# Making predictions on the test set
y_pred = svm_classifier.predict(X_test)

# Calculating the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
